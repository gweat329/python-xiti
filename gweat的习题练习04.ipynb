### 定义一个学生类，有下面的类属性
- 1.姓名
- 2.年龄
- 3.成绩(语文，数学，英语)每课成绩类型为整数类方法：
- 4.获取学生的姓名：get_name()返回类型:str
- 5.获取学生的年纪: get_age()返回类型:int
- 6.返回3门科目中最高的分数，get_course()返回类型:int

```python
class Student(object):
    def __init__(self, name, age, scores):
        self.name = name
        self.age = age
        self.scores = scores
    def get_name(self):
        return self.name
    def get_age(self):
        return self.age
    def get_course(self):
        return max(self.scores)
zz = Student("周周", 18, [86,76,50])
print(zz.get_name())
print(zz.get_age())
print(zz.get_course())
        
```

### 定义一个字典类：DictClass,完成如下功能
- 1.删除某个key del_dict(key)
- 2.判断某个键是否在字典里，如果在返回键对应的值，不在则返回'not found' get_dict()
- 3.返回键组成的列表 返回类型：list get_key()
- 4.合并字典，并且返回合并后字典的values组成的列表，返回类型list update_dict()

```python
class DictClass(object):
    
    def __init__(self,dict):
        self.dict = dict
    def del_dict(self,key):
        if key not in self.dict.keys():
            print("输入的不在字典里")
           
        else:
            self.dict.pop(key)
            return "删除成功"
    def get_dict(self,key):
        if key not in self.dict.keys():
            return "not found"
        else:
            return self.dict[key]
    def get_key(self):
        return self.dict.keys()
    def update_dict(self,dict2):
        self.dict = dict(self.dict,**dict2)
        return self.dict.values()
    
dict1 = DictClass({"a":1,"b":2,"c":3})
print(dict1.update_dict({"d":4}))
```

### 定义一个列表的操作类 ListInfo
#### 包括的方法
- 1.列表元素添加：add_key() ,添加的key必须是字符串或者数字
- 2.列表元素取值：get_key()
- 3.列表合并：update_list(list)
- 4.删除并且返回最后一个元素：del_key()

```python
class ListInfo(object):
    def __init__(self,list):
        self.list = list
    def add_key(self,ele):
        # 添加的key必须是字符串或者数字
        if isinstance(ele,(str,int)):
            self.list.append(ele)
            print(self.list)
            return "添加 {0} 成功".format(ele)
        else:
            return "添加的不是字符串和数字"
    def get_key(self,index):
        # 这里需要判断索引是否超出列表范围
        if index >= 0 and index < len(self.list):
            return self.list[index]
        else:
            print("超出列表取值范围")
    def update_list(self,list1):
        self.list.extend(list1)
        return self.list
    def del_key(self):
        if len(self.list) >= 0:
            return self.list.pop(-1)

        else:
            return "列表为空"

l = ListInfo([1,24,56,8,4])
print(l.add_key("我"))
#print(l.update_list([i for i in range(9)]))
#print(l.del_key())




        
```

### 创建北京和成都两个校区

- 管理员创建了讲师小周，并将其分配给了Python 3期
- 讲师小周创建了一条Python 3期的上课记录Day02
- 讲师小周为Day02这节课所有的学员批改了作业，小张得了A,小王得了B
- 学员小张查看了自己所报的课程
- 学员小张在查看了自己在Python 3的成绩列表然后退出了
- 学员小张给了讲师小周好评

```python
# 创建一个全局课程列表为空
course_list = []
# 创建一个学校类
class Schools(object):
    # 创建一个构造函数
    # 函数内有学校名字（为后面的北京校区和成都校区）
    def __init__(self,school_name):
        self.school_name = school_name
        # 学校内有老师和学生
        # 那么创建一个学生列表为空
        # 创建一个老师列表为空
        self.students_list = []
        self.teachers_list = []
        # 让变量course_list为全局变量
        global course_list
    # 登记学生，并增加至学生列表中
    def register_stu(self,stu):
        self.students_list.append(stu)
        
    # 聘请老师，并增加至学生列表中
    def employ_tea(self,tea):
        self.teachers_list.append(tea)
# 创建年级类（父类为Schools）
class Grade(Schools):
    # 定义构造函数，参数：self,school_name,grade_code,grade_course
    def __init__(self, school_name, grade_code, grade_course):
        # 引用super调用父类的构造函数__init__
        super(Grade, self).__init__(school_name)
        
        self.code = grade_code
        self.course = grade_course
        # 增加一个属性变量self.member,赋值一个空列表
        self.members = []
        # 一旦调用，对象课程增加至全局课程列表中
        course_list.append(self.course)
        print("我们现在有了{}的{}期的{}".format(school_name,self.code,self.course))
        # 定义一个课程信息
    def course_info(self):
        print("课程大纲{}是 day01, day02, day03".format(self.course))
# 实例化Grade一个对象Python和Linux
Python = Grade("北京","3","Python课程")
Linux = Grade("成都","1","Linux课程")
# 定义一个学校成员有哪些的类
class Schools_member(object):
    def __init__(self, name, age, sex,role):
        self.name = name
        self.age = age
        self.sex = sex
        self.role = role
        # 创建一个对象类属性self.course_list为空列表
        self.course_list = []
        
        print("我叫{0}，今年{1}岁，我是一个{2}。".format(self.name, self.age, self.role))
# 定义一个学生id，两位数       
stu_num_id = 00  
# 创建一个父类是学校成员的学生子类，引用父类属性，并增加一个父类没有的属性
class Students(Schools_member):
    def __init__(self,name, age, sex, role, course):
        super(Students, self).__init__(name, age, sex, role)
        # 全局声明学生id
        global stu_num_id
        # 学生id没引用一次自增一次
        stu_num_id += 1
        # 创建一个类属性并赋值
        stu_id = course.school_name + "S" + str(course.code) + str(stu_num_id).zfill(2)
        # 创建一个对象id，并给对象id赋值
        self.id = stu_id
        # 创建一个对象成绩列表为空的字典集合
        self.mark_list = {}
    # 定义学生学习的类方法
    def study(self, course):
        print("我来这里学习{0},我的学号是{1}。".format(course.course, self.id))
    # 定义交学费的类方法    
    def pay(self, course):
        print("我交了1000元的学费给{0}。".format(course.course))
        # 创建一个交学费的课程增加至对象课程列表的空列表中
        self.course_list.append(course.course)
    # 定义一个赞扬老师的类方法
    def praise(self, obj):
        print("{0}觉得{1}课真棒".format(self.name, obj.name))
    # 定义一个离开的类方法
    def out(self):
        print("我离开了")
# 创建一个老师id
tea_num_id = 00
# 创建一个父类是学校成员的老师子类，引用父类属性，并增加一个父类没有的属性
class Teachers(Schools_member):
    def __init__(self,name, age, sex, role, course):
        super(Teachers, self).__init__(name, age, sex, role)
        global tea_num_id
        tea_num_id += 1
        tea_id = course.school_name + "T" + str(course.code) + str(tea_num_id).zfill(2)
        self.id = tea_id
    
    def teach(self,course):
        print("我来这里讲{}门课，我的id是{}".format(course.course, self.id))\
            
    def record_mark(self, Date, course, obj, level):
        # 对象成绩字典形式的集合，并给 ‘键’ 赋 ‘值’
        obj.mark_list["Day"+Date] = level

a = Students("小陈", 20, "男", "学生", Python)
Python.register_stu(a)
a.study(Python)
a.pay(Python)
print("-"*30)
b = Students("小曾", 18, "女", "学生", Python)
Python.register_stu(b)
b.study(Python)
b.pay(Python)
print("-"*30)
t = Teachers("周老师", 30, "男", "老师", Python)
Python.employ_tea(t)
t.teach(Python)
t.record_mark("01", Python, a, "A")
t.record_mark("01", Python, b, "B")
t.record_mark("02", Python, a, "A")
t.record_mark("02", Python, b, "B")
print("小陈查看他的课程表")
print(a.course_list)
print("小陈查看他的成绩单")
print(a.mark_list)
print("小曾成绩不好退出了")
b.out()
print("小陈想给老师好评")
a.praise(t)


        

```
